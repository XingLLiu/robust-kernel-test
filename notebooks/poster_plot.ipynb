```python
import numpy as np
import scipy.stats as sci_stats
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0., 11., 1000)
figs, axes = plt.subplots(nrows=1, ncols=2, figsize=(8, 2))

p_norm = sci_stats.norm.pdf(x, loc=6.)
a, s = 2., 1.
p_gam = sci_stats.gamma.pdf(x, a=a, scale=s)

t = sci_stats.gamma.ppf(0.95, a=a, scale=s)
print("quantile:", t)
i = 431
print("emp quantile:", x[i])

ax = axes[0]
ax.plot(x, p_norm)
ax.plot(x, p_gam)
ax.fill_between(x[i:], p_norm[i:], color="C0", alpha=0.2)
ax.fill_between(x[i:], p_gam[i:], color="C1", alpha=0.4)
ax.axvline(x=x[i], color="grey", linestyle="--", linewidth=2)
ax.axis("off")

# 2.
p_norm = sci_stats.gamma.pdf(x, a=3.5, scale=1.)
a, s = 2., 1.
p_gam = sci_stats.gamma.pdf(x, a=a, scale=s)

t = sci_stats.gamma.ppf(0.95, a=a, scale=s)

ax = axes[1]
ax.plot(x, p_norm)
ax.plot(x, p_gam)
ax.fill_between(x[i:], p_norm[i:], color="C0", alpha=0.2)
ax.fill_between(x[i:], p_gam[i:], color="C1", alpha=0.4)
ax.axvline(x=x[i], color="grey", linestyle="--", linewidth=2)
ax.axis("off")

plt.tight_layout()
plt.savefig("../figs/poster.png")
```
